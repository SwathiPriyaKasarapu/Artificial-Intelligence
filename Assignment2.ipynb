```python
3)Best-First Search (BFS) and A* Search are both informed search algorithms that use heuristics to guide their pathfinding, but they 
differ significantly in how they use those heuristics and in their performance. Best-First Search selects the next node to explore 
based solely on the heuristic estimate of the cost to reach the goal (denoted as h(n)). This means it always chooses the node that
appears closest to the goal, often resulting in faster paths in simple scenarios but potentially ignoring the actual cost to reach 
that node, which can lead to inefficient or even incorrect paths.

In contrast, A* Search combines the heuristic estimate (h(n)) with the actual cost to reach the node (g(n)), evaluating nodes based
on f(n) = g(n) + h(n). This results in more balanced and often optimal pathfinding, especially when the heuristic is admissible 
(never overestimates the cost to the goal). While A* typically explores more nodes than Best-First Search and can be slower in practice, 
it is generally more reliable and accurate. In summary, Best-First Search is faster but less thorough, while A* is more computationally 
intensive but guarantees the shortest path when used with a proper heuristic.
```
